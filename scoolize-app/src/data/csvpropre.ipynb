```python
import pandas as pd
import os
```

```python
chemin_csv = 'scoolize-app/src/data/fr-esr-cartographie_formations_parcoursup - fr-esr-cartographie_formations_parcoursup.csv'
if not os.path.exists(chemin_csv):
    chemin_csv = 'fr-esr-cartographie_formations_parcoursup - fr-esr-cartographie_formations_parcoursup.csv'

df = pd.read_csv(chemin_csv)
df.head()
```

```python
colonnes = [
    'Identifiant de l\'établissement',
    'Nom de l\'établissement',
    'Types d\'établissement',
    'Commune',
    'Département',
    'Région'
]

df_etab = df[colonnes].copy()
df_etab.columns = ['code_uai', 'nom', 'statut', 'commune', 'departement', 'region']
```

```python
df_etab['nom'] = df_etab['nom'].str.strip()
df_etab = df_etab[df_etab['nom'].notna()]
df_etab = df_etab.drop_duplicates(subset=['code_uai'])
```

```python
lycees = df_etab[df_etab['nom'].str.contains('Lycée', case=False, na=False)].copy()
lycees = lycees[lycees['statut'].str.contains('Publics|Privés', case=False, na=False)]
```

```python
superieur = df_etab.copy()
```

```python
output_dir = 'scoolize-app/public/data'
if not os.path.exists(output_dir):
    output_dir = '../../public/data'

lycees.to_csv(f'{output_dir}/lycees.csv', index=False, encoding='utf-8', sep=';')
superieur.to_csv(f'{output_dir}/superieur_clean.csv', index=False, encoding='utf-8', sep=';')
```
